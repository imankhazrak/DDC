## Some imports.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

# Plots

```python

# data_folder = '/content/drive/MyDrive/4. BGSU/1.2 PhD_Courses/Z. RA & Desertation/Codes/Dr_Niu/MMD/Office31'
data_folder = r"/mnt/c/Users/Iman/Google Drive/4. BGSU/1.2 PhD_Courses/Z. RA & Desertation/Codes/Dr_Niu/MMD/Office31"
df_web = pd.read_csv(os.path.join(data_folder, 'test_train_result_amazon_webcam_resnet50.csv'))
df_dslr = pd.read_csv(os.path.join(data_folder, 'test_train_result_amazon_dslr_resnet50.csv'))
df_web_101 = pd.read_csv(os.path.join(data_folder, 'test_train_result_amazon_webcam_resnet101.csv'))
df_dslr_101 = pd.read_csv(os.path.join(data_folder, 'test_train_result_amazon_dslr_resnet101.csv'))

```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

for ax in axes.ravel():
    ax.set_ylim(0.7, 0.85)
#############################
axes[0,0].plot(df_web.Lambda, df_web['Test Accuracy'], 'ro-', label="Test Accuracy")
for x, y in zip(df_web.Lambda, df_web['Test Accuracy']):
    axes[0,0].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[0,0].set_title("Loss: MMD, resnet50,  \n train data: Amazon,test data: webcam \n Train and Test accuracy vs Lambda")
axes[0,0].set_xlabel("Lambda")
axes[0,0].set_ylabel("Accuracy")

y_values = df_web.groupby('Lambda')['Accuracy'].last()
axes[0,0].plot(df_web.Lambda.unique(), y_values, 'o-', label="Train Accuracy")
for x, y in zip(df_web.Lambda.unique(), y_values):
    axes[0,0].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[0,0].legend()

##############################
axes[0,1].plot(df_dslr.Lambda, df_dslr['Test Accuracy'], 'ro-', label="Test Accuracy")
for x, y in zip(df_dslr.Lambda, df_dslr['Test Accuracy']):
    axes[0,1].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[0,1].set_title("Loss: MMD, resnet50,  \n train data: Amazon,test data: dslr \n Train and Test accuracy vs Lambda")
axes[0,1].set_xlabel("Lambda")
axes[0,1].set_ylabel("Accuracy")

y_values = df_dslr.groupby('Lambda')['Accuracy'].last()
axes[0,1].plot(df_dslr.Lambda.unique(), y_values, 'o-', label="Train Accuracy")
for x, y in zip(df_dslr.Lambda.unique(), y_values):
    axes[0,1].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[0,1].legend()

#############################
axes[1,0].plot(df_web_101.Lambda, df_web_101['Test Accuracy'], 'ro-', label="Test Accuracy")
for x, y in zip(df_web_101.Lambda, df_web_101['Test Accuracy']):
    axes[1,0].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[1,0].set_title("Loss: MMD, resnet101, \n train data: Amazon,test data: webcam \n Train and Test accuracy vs Lambda")
axes[1,0].set_xlabel("Lambda")
axes[1,0].set_ylabel("Accuracy")

y_values = df_web_101.groupby('Lambda')['Accuracy'].last()
axes[1,0].plot(df_web_101.Lambda.unique(), y_values, 'o-', label="Train Accuracy")
for x, y in zip(df_web_101.Lambda.unique(), y_values):
    axes[1,0].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[1,0].legend()

##############################
axes[1,1].plot(df_dslr_101.Lambda, df_dslr_101['Test Accuracy'], 'ro-', label="Test Accuracy")
for x, y in zip(df_dslr_101.Lambda, df_dslr_101['Test Accuracy']):
    axes[1,1].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[1,1].set_title("Loss: MMD, resnet101, \n train data: Amazon,test data: dslr \n Train and Test accuracy vs Lambda")
axes[1,1].set_xlabel("Lambda")
axes[1,1].set_ylabel("Accuracy")

y_values = df_dslr_101.groupby('Lambda')['Accuracy'].last()
axes[1,1].plot(df_dslr_101.Lambda.unique(), y_values, 'o-', label="Train Accuracy")
for x, y in zip(df_dslr_101.Lambda.unique(), y_values):
    axes[1,1].annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')
axes[1,1].legend()


plt.tight_layout()
plt.show()
```
